# Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.
Probability Mass Function (PMF):
The PMF is used for discrete random variables and gives the probability that a discrete random variable is exactly equal to some value.

Probability Density Function (PDF):
The PDF is used for continuous random variables and represents the likelihood of a random variable taking a value within a given range. The probability that a variable falls within an interval is the integral of the PDF over that interval.


# Q2: What is Cumulative Distribution Function (CDF)? Explain with an example. Why is the CDF used?
Cumulative Distribution Function (CDF):
The CDF is used to describe the cumulative probability that a random variable X takes a value less than or equal to a given value x. It is defined for both discrete and continuous random variables and is the integral (sum for discrete) of the PDF/PMF.

Why is CDF used?: The CDF is useful because it allows for the calculation of the probability that a variable falls within a specific range (i.e., between two values). It helps to compare different distributions and provides a complete description of the distribution of a random variable.

# Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution.
Importance of Normal Distribution:
The normal distribution is important because it is widely used in statistics due to its properties and the fact that many natural and social phenomena tend to follow it.
It is also fundamental in inferential statistics for hypothesis testing, confidence intervals, and regression analysis.
Real-life Examples:
IQ Scores: IQ scores in the population are often modeled as a normal distribution with a mean of 100 and a standard deviation of 15.

# Q9: What is the Central Limit Theorem? State the significance of the Central Limit Theorem.

Central Limit Theorem (CLT):
The CLT states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables approaches a normal distribution, regardless of the original distribution of the variables, as the sample size increases.

Significance:

The CLT allows us to apply normal distribution techniques to sample means, even if the population distribution is not normal.
It forms the basis for many statistical tests and confidence intervals.
